```
#to study more about VADER, move to https://medium.com/analytics-vidhya/simplifying-social-media-sentiment-analysis-using-vader-in-python-f9e6ec6fc52f

nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```
#now get rid of any links, mentions or any special characters so that we are left with only words that can be checked for sentiment analysis

import re

def clean_tweet(tweet):
  
  return ' '.join( re.sub("(@[A-Za-z0-9]+)|(^0-9A-Za-z \t)|(\w+:\/\/\S+)",' ',tweet).split())
```

```
#now store the cleaned tweets in the dataframe

data = pd.DataFrame(data = [clean_tweet(tweet.text) for tweet in tweets], columns = ["Tweets"])
```

```
tweets[0].id
```

```
#now perform the analysis

listy = []

for index, row in data.iterrows():
  ss = sia.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
data['polarity'] = se.values

for index,row in data.iterrows():
  if data.iloc[index]['polarity']['compound']>=0.5:
    print(row["Tweets"])
    print("Positive Sentiment")
  
  elif data.iloc[index]['polarity']['compound']> -0.5 and data.iloc[0]['polarity']['compound']<0.5:
    print(row["Tweets"])
    print("Neutral Sentiment")
  
  else:
    print(row["Tweets"])
    print("Negative Sentiment")    
    
  print("\n")
```
